# Gráficos de Línea

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 50)
y = 2*x-15
plt.style.use('ggplot')
plt.plot(x,y)
plt.title("Función lineal")
```

```python
x = np.linspace(0, 20)
y = x**2-3*x-35

plt.plot(x,y,color="r")
plt.title("Función cuadrática")
```

```python
x = np.linspace(0, 10)
y = np.exp(x) 

plt.plot(x,y,color="g")
plt.title("Función exponencial")
```

```python
x = np.linspace(0, 40, 400)
y = np.sin(x) 

plt.plot(x,y,color="m")
plt.title("Función senoidal")
```

```python
from random import randint
x = range(1,31)
y = [randint(3500, 4000) for i in range(30)]
std = np.std(y)
var = np.var(y)
plt.plot(x,y)
plt.ylim(0,4000)
plt.xlabel("Días")
plt.ylabel("Dólares ($)")
plt.xlim(1, 30)
plt.title(f"Ingresos del mes de junio - iStore    std: {std:.1f}  var: {var:.1f}")
```

# Gráficos de barras

```python
import matplotlib.pyplot as plt
import numpy as np

x = ["Enero", "Febrero", "Marzo", "Abril", "Mayo", "Junio"]
y = [11.5, 14.8, 9.6, 13.2, 15.9, 16.5]
print(f"Media: {np.array(y).mean():.1f}")
print(f"std: {np.array(y).std():.2f}")
print(f"var: {np.array(y).var():.2f}")
plt.bar(x,y)
plt.title("Ventas en la primera mitad del 2022")
plt.ylabel("Miles de dólares")
```

```python

# Graficando histograma
mu, sigma = 0, 0.2 # media y desvio estandar
datos = np.random.normal(mu, sigma, 1000) #creando muestra de datos
std = np.std(datos)
var = np.var(datos)
print(f"Media: {datos.mean():.2f}")
print(f"std: {np.std(datos):.2f}")
print(f"var: {np.var(datos):.2f}")
# histograma de distribución normal.
cuenta, cajas, ignorar = plt.hist(datos, 20, color="b")
plt.ylabel('frequencia')
plt.xlabel('valores')
plt.title('Histograma')
```

```python
import matplotlib.pyplot as plt

from sklearn import datasets
from sklearn.decomposition import PCA

iris = datasets.load_iris()

X = iris.data
y = iris.target
target_names = iris.target_names

pca = PCA(n_components=2)
X_r = pca.fit(X).transform(X)

colors = ['navy', 'turquoise', 'darkorange']
lw = 2

for color, i, target_name in zip(colors, [0, 1, 2], target_names):
    plt.scatter(X_r[y == i, 0], X_r[y == i, 1], color=color, alpha=.8, lw=lw,
                label=target_name)
plt.legend(loc='best', shadow=False, scatterpoints=1)
plt.title('PCA IRIS dataset')
```

## Gráfico de áreas apiladas

```python
import matplotlib.pyplot as plt 

data1=[35,20,10,5,5]
data2=[25,20,20,15,5]
data3=[40,50,70,80,90]

year=["2015","2016","2017","2018","2019"]

fig,ax=plt.subplots(3,1,figsize=(9,7))
ax[0].set_title("Lenguajes utilizados en proyectos de Google")
ax[0].bar(year,data1)
ax[0].legend(["C++"])
ax[1].bar(year,data2)
ax[1].legend(["JavaScript"])
ax[2].bar(year,data3)
ax[2].legend(["Python"])
```

```python
import pandas as pd
import matplotlib.pyplot as plt

years=["2015","2016","2017","2018","2019"]
data={"Python":[40,50,70,80,90], "JavaScript":[25,20,20,15,5], "C++":[35,30,10,5,5]}

df=pd.DataFrame(data,index=years)
plt.style.use('ggplot')
df.plot(kind="bar",stacked=True,figsize=(9,6))
plt.legend(loc="lower left",bbox_to_anchor=(0.8,1.0))
plt.title("Lenguajes utilizados en proyectos de Google")
```

```python
df=pd.DataFrame(data,index=years)
plt.style.use('ggplot')
df.plot.barh(stacked=True,figsize=(9,6))
plt.legend(loc="lower left",bbox_to_anchor=(0.8,1.0))
plt.title("Lenguajes utilizados en proyectos de Google")
```

## Gráfico de histograma

```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('ggplot')

datos = np.random.uniform(0, 1.5, 10000)

plt.hist(datos, bins=25, color="g")
plt.title("Distribución Uniforme")
plt.ylabel("Frecuencia")
plt.show()
```

```python
datos = np.random.normal(0, 1.5, 10000)

plt.hist(datos, bins=25, color="m")
plt.title("Distribución Gaussiana (normal)")
plt.ylabel("Frecuencia")
plt.show()
```

```python
datos = np.random.gamma(5, 1, 10000)

plt.hist(datos, bins=25, color="c")
plt.title("Distribución Gamma")
plt.ylabel("Frecuencia")
plt.show()
```

```python
datos = np.random.beta(3, 2, 10000)

plt.hist(datos, bins=25, color="y")
plt.title("Distribución Beta")
plt.ylabel("Frecuencia")
plt.show()
```

```python
datos = np.random.laplace(0, 1.5, 10000)

plt.hist(datos, bins=50, color="b")
plt.title("Distribución Laplaciana")
plt.ylabel("Frecuencia")
plt.show()
```

```python
datos = np.random.exponential(10, 10000)

plt.hist(datos, bins=40)
plt.title("Distribución Exponencial")
plt.ylabel("Frecuencia")
plt.show()
```
