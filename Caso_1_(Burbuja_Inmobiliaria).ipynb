```python
import pandas as pd
import plotly.graph_objects as go

cities = [
    {"city": "Miami",       "rank": 1,  "score": 1.7, "region": "Américas"},
    {"city": "Tokyo",       "rank": 2,  "score": 1.6, "region": "Asia-Pacífico"},
    {"city": "Zurich",      "rank": 3,  "score": 1.6, "region": "Europa"},
    {"city": "Los Angeles", "rank": 4,  "score": 1.1, "region": "Américas"},
    {"city": "Dubai",       "rank": 5,  "score": 1.1, "region": "Medio Oriente"},
    {"city": "Amsterdam",   "rank": 6,  "score": 1.1, "region": "Europa"},
    {"city": "Geneva",      "rank": 7,  "score": 1.1, "region": "Europa"},
    {"city": "Toronto",     "rank": 8,  "score": 0.8, "region": "Américas"},
    {"city": "Sydney",      "rank": 9,  "score": 0.8, "region": "Asia-Pacífico"},
    {"city": "Madrid",      "rank": 10, "score": 0.8, "region": "Europa"},
]

color_map = {
    "Américas":      "#E63946",
    "Europa":        "#457B9D",
    "Asia-Pacífico": "#2A9D8F",
    "Medio Oriente": "#E9C46A",
}

df = pd.DataFrame(cities)
df_sorted = df.sort_values("score", ascending=False).reset_index(drop=True)
df_sorted["label"] = df_sorted["city"] + " (#" + df_sorted["rank"].astype(str) + ")"

fig = go.Figure()

for region in df_sorted["region"].unique():
    grp = df_sorted[df_sorted["region"] == region]
    fig.add_trace(go.Bar(
        y=grp["label"],
        x=grp["score"],
        name=region,
        orientation="h",
        marker_color=color_map[region],
        marker_opacity=0.88,
        text=grp["score"].apply(lambda v: f"{v:.1f}"),
        textposition="outside",
    ))

fig.update_layout(
    # ── Título centrado + subtítulo ──────────────────────────────────────────
    title=dict(
        text=(
            "<b>Riesgo de burbuja inmobiliaria — ranking por ciudad</b><br>"
            "<sup style='color:gray'>Puntuación UBS: >1.5 = burbuja, 0.5–1.5 = riesgo elevado, <0.5 = moderado</sup>"
        ),
        x=0.5,           # centrado
        xanchor="center",
        y=0.97,
        font=dict(size=14),
    ),

    # ── Leyenda en esquina superior derecha ──────────────────────────────────
    legend=dict(
        orientation="v",
        x=1.0,
        y=1.0,
        xanchor="right",
        yanchor="top",
        title_text="Región",
        bgcolor="rgba(255,255,255,0.8)",
        bordercolor="#e0e0e0",
        borderwidth=0.5,
    ),

    # ── Ejes ─────────────────────────────────────────────────────────────────
    #xaxis=dict(
     #   title="Puntuación de riesgo",
      #  gridcolor="#e8e6df",
       # gridwidth=0.5,
        #range=[0, 2.3],
        #zeroline=False,
    #),
    #yaxis=dict(
     #   title="",
      #  categoryorder="array",
       # categoryarray=df_sorted["label"].tolist()[::-1],
    #),

    # ── Fuente al pie ─────────────────────────────────────────────────────────
    annotations=[dict(
        text="Fuente: UBS Global Real Estate Bubble Index. Nota: Puntuación compuesta — precios vs. ingresos, alquileres, crédito y construcción.",
        xref="paper", yref="paper",
        x=0, y=-0.13,
        xanchor="left",
        showarrow=False,
        font=dict(size=9, color="gray"),
    )],

    barmode="group",
    height=460,
    margin=dict(t=90, b=70, l=10, r=160),
    plot_bgcolor="white",
    paper_bgcolor="white",
)

fig.show()
```
